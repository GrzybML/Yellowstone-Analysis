# Yellowstone

Created by Taylor Sheridan and John Linson, the show stars Kevin Costner as John Dutton, who controls the largest contiguous ranch in the United States. The series is known for its complex characters and dramatic depiction of the conflicts at the borders of a large cattle ranch, an Indian reservation, and land developers.

## Viewership

### Viewership Graph

```python
import matplotlib.pyplot as plt

df['Cumulative Episode'] = range(1, len(df) + 1)

plt.figure(figsize=(14, 7))
plt.plot(df['Cumulative Episode'], df['Viewers'], marker='o', linestyle='-', color='b')
plt.title('Viewership Trend for Yellowstone Across All Seasons')
plt.xlabel('Cumulative Episode Number')
plt.ylabel('Viewers (in millions)')
plt.grid(True)
plt.show()
```

The viewership trend for "Yellowstone" shows a general increase over time, with significant peaks at season premieres and finales. Season 4 marks the highest viewership, suggesting growing popularity and a solid fan base. Despite some fluctuations, the series demonstrates strong viewer retention and interest across its seasons.

```python
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(10, 20), sharex=False)

for index, (season, group) in enumerate(df.groupby('Season')):
    axs[index].plot(group['Episode'], group['Viewers'], marker='o', color='tab:blue', linestyle='-', label=f'Season {season}')
    axs[index].set_title(f'Season {season} Viewership', fontsize=14)
    axs[index].set_ylabel('Viewers (millions)', fontsize=12)
    axs[index].grid(True)
    axs[index].legend(loc='upper right')
    axs[index].set_xticks(group['Episode']) 
    axs[index].set_xlabel('Episode Number')  
    
plt.tight_layout()
plt.show()
```
